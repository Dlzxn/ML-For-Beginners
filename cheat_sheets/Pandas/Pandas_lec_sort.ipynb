**Замена значений**
--

```python
df["sex"].replace('Female', "f")
# или
df["sex"].replace(["Female", "Male"], ["F", "M"]) # первый список исходных, второй - целевых
```

**Диапазоны**

```python
el = df["total_bill"].between(10, 20, inclusive='both') # попадают ли значения в диапазон
# inclusive отвечает за то, используем ли мы верхнюю границу
el
```

```python
df[el]
```

**Сортировка по столбцу с конкретной размерностью**
--

```python
df.nlargest(10, 'tip') # 10 самых больших значений в колонке tip
```

```python
df.nsmallest(10, 'tip') # 10 самых маленьких значений в колонке tip
```

**n случайных строк**
